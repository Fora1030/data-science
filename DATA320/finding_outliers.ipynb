```python
import pandas as pd

countries_df = pd.read_csv(r'c:\Users\Franklin\Desktop\local_repos\data-science\DATA320\week_3\surgical_procedures.txt', sep=',')
countries_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data = countries_df[["Country", "Weekly_Budget_USD"]]

# Create a figure with a specified size
fig = plt.figure(figsize=(10, 8))

# Create a GridSpec object
gs = fig.add_gridspec(2, 1, height_ratios=[1, 4])

# Create the box plot in the first subplot
ax1 = fig.add_subplot(gs[0])
sns.boxplot(data=data, ax=ax1, orient='h')
ax1.set(xlabel='')

# Create the histogram in the second subplot
ax2 = fig.add_subplot(gs[1])
sns.histplot(data, kde=False, ax=ax2)

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
df = countries_df[['Country', 'Weekly_Budget_USD']]

# Melt the DataFrame to long format
df_melted = df.melt(id_vars='Country', var_name='Metric', value_name='Value')

# Create a figure with a specified size
plt.figure(figsize=(12, 8))

# Create a box plot for each metric
sns.boxplot(x='Metric', y='Value', hue='Country', data=df_melted)

# Set the title and labels
plt.title('Box Plot of Metrics by Country')
plt.xlabel('Metric')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python

# Set the figure size
df = countries_df
plt.figure(figsize=(15, 10))

# Loop through each column and create a box plot
for i, column in enumerate(df.columns[1:], 1):  # Skip the 'Country' column
    plt.subplot(1, len(df.columns) - 1, i)
    sns.boxplot(y=df[column])
    plt.title(column)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
