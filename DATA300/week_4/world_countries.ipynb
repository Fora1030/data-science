# EDA Case Study -- World Countries

```python
%matplotlib inline
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sbn
import altair as alt
from altair import Chart, X, Y, Color, Scale
from vega_datasets import data
import requests
from bs4 import BeautifulSoup
matplotlib.style.use('ggplot')
```

```python
wb = pd.read_csv('world_countries.csv')
```

```python
c = Chart(wb)
# make a chart
m = c.mark_bar() # set the mark -- returns a new Chart
e = m.encode(X('Birthrate',bin=True),y='count()') # set the encoding
e.display()
```

```python
Chart(wb).mark_bar().encode(x=X('Birthrate', bin=True), y='count()')
```

```python
Chart(wb).mark_point().encode(x='Population', y='Area', tooltip='Country')
```

```python
alt.Chart(wb).mark_circle().encode (
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='Region:N'
).properties(
    width=150,
    height=150
).repeat(
    row=['Birthrate', 'Deathrate', 'Infant mortality', 'GDP'],
    column=['Birthrate', 'Deathrate', 'Infant mortality', 'GDP']
).interactive()
```
