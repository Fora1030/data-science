```python
import pandas as pd
import matplotlib.pyplot as plt

gapminder_gdp_asia = pd.read_csv('gapminder_gdp_asia.csv', sep=',')
fig, ax = plt.subplots()
gapminder_gdp_asia.set_index("country").T.plot(ax=ax,kind='line', y=["China", "India"])
plt.xticks(rotation=45)
lbls = [str(i).split('_')[1] for i in gapminder_gdp_asia.T.index[1:]]
plt.xticks(ticks=range(len(gapminder_gdp_asia.T.index)-1), labels=lbls)
plt.title("China vs India GDP from 1952 to 2007")
plt.legend(loc='lower right')
plt.xlabel("Year")
plt.ylabel("GDP")
plt.show()
```
