```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
# import the csv file and parse the DATE column as date dtype and set it as the index
data = pd.read_csv(r'helsinki-vantaa.txt', sep='\t', parse_dates=['DATE'], index_col="DATE")
data.sort_index(inplace=True) # sort the data
selection = data.loc['1988-01-01':'2018-12-01'] # filter the 30 years
fig, ax = plt.subplots()
selection.plot(kind='line', y=["TEMP_C"], ax=ax, style='-ok') # plot
fig.set_size_inches(14, 6)
plt.ylabel("Temperature (Celsius)")
plt.xlabel("Time")
plt.xticks(rotation=45)
plt.title("Helsinki-Vantaa Airport")
plt.show()
```
